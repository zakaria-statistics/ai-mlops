```python
def fit(X,y):
    lr=0.001
    epochs=10000
    bias=0.5
    w1=0.5
    n=len(y)
   

    for i in range(epochs):
        py=bias+w1*X
     
        dbias=-(2/n)*sum(y-py)
        dw1=-(2/n)*sum(X*(y-py))
          
        w1=w1-dw1*lr
        bias=bias-dbias*lr
    return bias,w1
```

```python
#x1:H(X1)=ax1+b
#H(3)=0.1*3+0.7=1=predicted value for x1=3
#actual value for x1=3
```

```python
def linear(z):
    return z
class NN1Neurone:
    def __init__(self):
        self.lr=0.001
        self.epochs=10000

    def fit(self,X,y):
        self.bias=0.7
        self.w1=0.1
        n=len(y)
        for i in range(self.epochs):
            ws=self.bias+self.w1*X
            py=linear(ws)
            

            dbias=-(2/n)*sum(y-py)
            dw1=-(2/n)*sum(X*(y-py))

            self.w1=self.w1-dw1*self.lr
            self.bias=self.bias-dbias*self.lr
    def predict(self,X):
        ws=self.bias+self.w1*X
        return linear(ws)
    def getDetail(self):
        return (self.bias,[self.w1])
```

```python
nn1=NN1Neurone()
nn1.fit(X_train,y_train)
nn1.getDetail()
```

```python
yhat=nn1.predict(X_test)
mse=mean_squared_error(y_test,yhat)
print(mse)
```
